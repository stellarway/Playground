```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://search.naver.com/search.naver'
keyword = 'nlp'
params = {'where':'news', 'query': keyword}
resp = requests.get(url, params = params)
```

```python
resp.content
```

```python
soup = BeautifulSoup(resp.content)
```

```python
num = soup.find('div', class_= 'title_desc all_my').text.split(' ')[-1].replace('','').replace(',','')
```

```python
ul_list=soup.find('ul', class_='type01').findAll('dt')

for ul in ul_list:
    title=ul.find('a')['title']
    print(title)
```

```python
news_list = []
ul_list=soup.find('ul', class_='type01').findAll('dl')

for ul in ul_list:
    title=ul.find('dt').find('a')['title']
    link=ul.find('dt').find('a')['href']
    news_dict={
        'title':title,
        'link':link
    }
    news_list.append(news_dict)
```

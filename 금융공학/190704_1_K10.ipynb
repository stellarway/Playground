```python
import requests
import pandas as pd
import xml.etree.ElementTree as ET
```

```python
def getStockPrice(StockCode, Count = 252) :
#     ret = pd.DataFrame(columns=['date','close'])
    ret = pd.DataFrame(columns=[StockCode])
    url = "https://fchart.stock.naver.com/sise.nhn?symbol={}&timeframe=day&count={}&requestType=0"
    url = url.format(StockCode, Count)
    response = requests.get(url)
    e = ET.fromstring(response.text.replace('\n','').replace('\t',''))
    
    for item in e.find('chartdata').findall('item') :
#         date = item.get('data').split('|')[0]
        close = int(item.get('data').split('|')[4])
      
        ret.loc[len(ret)] = close
   
    return ret
```

```python
# K10_cd=['005930', '000660', '068270', '207940', '005380', '005490', '051910', '028260', '035420', '012330']
```

```python
df_cls=pd.DataFrame()
for item in K10_cd:
    df_cls=df_cls.join(getStockPrice(item), how='outer')
#     df_total=df_total.merge(getStockPrice(item),left_index=True, right_index=True, how='outer')
```

### 실습1. 로그 수익률 계산
#### 1. 삼성전자 수정주가 조회 함수 구현

```python
import requests
import pandas as pd
import numpy as np
import xml.etree.ElementTree as ET
```

```python
def getStockPrice(StockCode='005930', Count = 252):
    ret = pd.DataFrame(columns=['date', 'open', 'high', 'low', 'close','volumn'])
    
    url="https://fchart.stock.naver.com/sise.nhn?symbol={}&timeframe=day&count={}&requestType=0"
    url= url.format(StockCode, Count)
    response= requests.get(url)
 
    e= ET.fromstring(response.text.replace('\n','').replace('\t',''))
   
    for item in e.find('chartdata').findall('item'):
        data = item.get('data').split('|')
        ret.loc[len(ret)] = data
        
    return ret
```

```python
ss=getStockPrice()
```

### 실습1. 로그 수익률 계산
#### 2. 삼성전자 로그 수익률 계산 함수 구현

```python
def get_yield(df):
    df['yield']=np.NaN
    for n in range(1,len(df)):
        df['yield'][n]=np.log(float(df['close'][n-1])/float(df['close'][n]))
    return df
```

```python
ss_y=get_yield(ss)
ss_y
```

### 실습2. 로그 수익률 평균
#### 1. 로그수익률 평균 함수 구현
입력: 로그수익률 데이터프레임  
출력: 평균= 로그수익률 합/데이터건수


```python
def get_mean(df):
    mean=df['yield'].sum()/df['yield'].count()
    return mean
```

```python
mean=get_mean(ss_y)
mean
```

### 실습3. 표준편차
#### 1. 표준편차 함수 구현
입력: 로그수익률 데이터프레임  
출력: 표준편차

```python
ss_y.describe()
```

```python
def get_std(df):
    std=np.sqrt(((df['yield']-mean)**2).sum()/df['yield'].count())
    return std
```

```python
std=get_std(ss_y)
std
```

### 실습3. 표준편차
#### 2. 변동성 함수 구현
입력: 로그수익률  
출력: 변동성

```python
def get_Var(df):
    Var=np.sqrt(((df['yield']-mean)**2).sum()/df['yield'].count())*np.sqrt(252)
    return Var
```

```python
Var=get_Var(ss_y)
Var
```

### 실습3. 표준편차
#### 3. 종목별 변동성 구현 (예. 005930)
입력: 로그수익률  
출력: 변동성

```python
def get_Var_from_code(StockCode, Count = 252) :
    # 수정주가 가져오기
    df = pd.DataFrame(columns=['close'])
    url = "https://fchart.stock.naver.com/sise.nhn?symbol={}&timeframe=day&count={}&requestType=0"
    url = url.format(StockCode, Count)
    response = requests.get(url)
    e = ET.fromstring(response.text.replace('\n','').replace('\t',''))
    for item in e.find('chartdata').findall('item') :
        close = item.get('data').split('|')[4]
        df.loc[len(df)] = close
    # 로그수익률 계산
    df['yield']=np.NaN
    for n in range(1,len(df)):
        df['yield'][n]=np.log(float(df['close'][n-1])/float(df['close'][n]))
    # 변동성 구현
    var=np.sqrt(((df['yield']-mean)**2).sum()/df['yield'].count())*np.sqrt(252)
    return var
    
```

### 실습4. 공분산, 상관계수
#### 2. 상관계수 계산 함수 구현
입력: 종목1로그수익률, 종목2로그수익률  
출력: 상관계수

```python
def get_corr(df1,df2):
    mean1=df1['yield'].sum()/df1['yield'].count()
    mean2=df2['yield'].sum()/df2['yield'].count()
    cov = ((df1['yield']-mean1)*(df2['yield']-mean2)).sum()/df1['yield'].count()
    
    std1=np.sqrt(((df1['yield']-mean1)**2).sum()/df1['yield'].count())
    std2=np.sqrt(((df2['yield']-mean2)**2).sum()/df2['yield'].count())
    
    corr=cov/(std1*std2)
    return corr
```

```python
get_corr(y1,y2)
```

### 실습4. 공분산, 상관계수
#### 3. 베타 께산 함수 구현
입력: 종목1로그수익률, 종목2로그수익률  
출력: 베타

```python
def get_beta(df1, df2):
    mean1=df1['yield'].sum()/df1['yield'].count()
    mean2=df2['yield'].sum()/df2['yield'].count()
    cov = ((df1['yield']-mean1)*(df2['yield']-mean2)).sum()/df1['yield'].count()
    
    Var2=np.sqrt(((df2['yield']-mean2)**2).sum()/df2['yield'].count())*np.sqrt(252)
    
    beta=cov/Var2
    return beta
    
```

```python
get_beta(y1,y2)
```

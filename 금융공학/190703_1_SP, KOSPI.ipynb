```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from urllib.request import urlopen
import json
```

```python
date_list=[]
price_list=[]
for i in range(1, 10):
    SP_url= 'https://finance.naver.com/world/worldDayListJson.nhn'
    symbol= 'SPI@SPX'
    params = {
        'symbol' : symbol,
        'fdtc' : 0,
        'page' : i
    }
    SP_resp=requests.get(SP_url, params)
    SP_json=SP_resp.json()
    for item in SP_json:     
        date= pd.to_datetime(item['xymd']).date()
        date_list.append(date)
        price = float(item['clos'])
        price_list.append(price)
       
        SP_df=pd.DataFrame({'S&P500':price_list}, index=date_list)
```

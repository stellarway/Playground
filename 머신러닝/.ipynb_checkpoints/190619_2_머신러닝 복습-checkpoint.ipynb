```python

from sklearn.datasets import make_regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split

X, y = make_regression(n_samples=100, n_features=10, bias=10, noise=100)

X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True)
```

```python

# Ridge Regression 클래스 객체 생성
model = Ridge(alpha=0.1) 
# fit 메서드로 모형 추정 (오그멘테이션은 자동으로 수행)
model_lr = model.fit(X_train, y_train)

print('weight: ', model_lr.coef_)
print('bias: ', model_lr.intercept_) # W0를 의미
```

```python

model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
print(type(model))
```

```python
print(type(model.fit))
```

##  보스턴 데이터셋

```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python

# 데이터 로드
boston = load_boston()
# CRIM: 범죄율
# INDUS: 비소매상업지역 면적 비율
# NOX: 일산화질소 농도
# RM: 주택당 방 수
# LSTAT: 인구 중 하위 계층 비율
# B: 인구 중 흑인 비율
# PTRATIO: 학생/교사 비율
# ZN: 25,000 평방피트를 초과 거주지역 비율
# CHAS: 찰스강의 경계에 위치한 경우는 1, 아니면 0
# AGE: 1940년 이전에 건축된 주택의 비율
# RAD: 방사형 고속도로까지의 거리
# DIS: 직업센터의 거리
# TAX: 재산세율
boston.keys()
```

```python
boston['feature_names'] # = boston.feature_names
```

```python
X = pd.DataFrame(boston.data, columns=boston.feature_names)
y = pd.DataFrame(boston.target, columns=['price'])
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
```

```python
# Linear Regression 클래스 객체 생성
model = LinearRegression() # fit_intercept = 상수항 유무 결정
```

```python
# fit 메서드로 모형 추정 (오그멘테이션은 자동으로 수행)
model_lr = model.fit(X_train, y_train)
```

```python

print('weight: ', model_lr.coef_)
print('bias: ', model_lr.intercept_)
```

```python
model.fit(X, y)
```

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
```

```python

X, y = make_regression(n_samples=30, n_features=30, bias=10, noise=100)

X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True)
```

```python
# Ridge Regression 클래스 객체 생성
model = Ridge(alpha=0.1) # cost 계산 식에서 lamda (보정의 가중치)
# fit 메서드로 모형 추정 (오그멘테이션은 자동으로 수행)
model_lr = model.fit(X_train, y_train)

print('weight: ', model_lr.coef_)
print('bias: ', model_lr.intercept_)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model=Ridge(alpha=0.01)
model.fit(X_train, y_train)
print("%s  %s"%(model.score(X_train, y_train),
                model.score(X_test, y_test)))
```

```python
sorted_coef=np.sort(model.coef_)
idx=np.arange(len(model.coef_))
plt.bar(idx, sorted_coef)
plt.show()
```

```python
import numpy as np

np.random.seed(42) # 항상 같은 테스트 셋이 나오도록 

def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))  # 인덱스 셔플 (0~data의 길이만큼 random하게)
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size] # test_set_size까지. random 배열을 가져옴.
    train_indices = shuffled_indices[test_set_size:] # 나머지. -----> 굳이 구현할 필요 없이\ 
                                                    # sklearn에서 구현해둔 함수가 있으니 그걸 이용하자
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python

train_set, test_set = split_train_test(housing, 0.2) # 해당 데이터를 0.2 만큼 추출한다.
print(len(train_set), 'train +', len(test_set), 'test')
```

```python

# 계층적 샘플링을 위한 income_cat
housing['income_cat'] = np.ceil(housing['median_income'] / 1.5)  # income의 카테고리 수 제한

# where절 사용법: 
# where(cond, other, inplace): cond가 True => 원래값 유지, cond가 False면 other로 설정 함. inplace가 True면 중복값 덮어쓰기.
housing['income_cat'].where(housing['income_cat'] < 6, 6.0, inplace=True) # 6보다 큰 카테고리는 6로 합치자. (소득 카테고리를 나눔)
housing.head(10)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit # 계층 추출을 쉽게 만들어주는 class
```

```python
# StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42) 
# random_state는 random seed 의미
# split.split => generator
for train_idx, test_idx in split.split(housing, housing['income_cat']):
    strat_train_set = housing.loc[train_idx]
    strat_test_set = housing.loc[test_idx]
```

# sklearn을 이용한 TF-IDF

```python
d1 = "The cat sat on my face. I hate a cat."
d2 = "The dog sat on my bed. I love a dog."
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import defaultdict
```

```python
document_ls = [d1, d2]
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(document_ls)
```

```python
import pandas as pd
count_vect_df = pd.DataFrame(tfidf.todense(), columns = vectorizer.get_feature_names())
count_vect_df
```

```python
import numpy as np
feature_array = np.array(vectorizer.get_feature_names())
tfidf_sorting = np.argsort(tfidf[0].toarray()).flatten()[::-1]
n=3
top_n = feature_array[tfidf_sorting][:n]

top_n
```

```python
import nltk
import string
from nltk.tokenize import TreebankWordTokenizer as tree
nltk.download('punkt')
```

```python
text = tree().tokenize(Text)
print("Tokenized Text : \n")
print(text)
```

```python
nltk.download('averaged_perceptron_tagger')
POS_tag = nltk.pos_tag(text)
print(POS_tag)
```

```python
from nltk.corpus import wordnet
nltk.download('wordnet')
```

```python
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

lemmatized_text = []

for word in POS_tag:
    lemmatized_text.append(str(wordnet_lemmatizer.lemmatize(word[0], pos = get_wordnet_pos(word[1]))))
print(lemmatized_text)
```

```python
stopwords = []
wanted_POS = ['NN', 'NNS', 'NNP', 'NNPS', 'JJ', 'JJR', 'JJS']

for word in POS_tag:
    if word[1] not in wanted_POS:
        stopwords.append(word[0])

punctuations = list(string.punctuation)
stopwords = stopwords + punctuations

stopwords_plus = ['t', 'isn']
stopwords = stopwords + stopwords_plus
stopwords = set(stopwords)

processed_text = []
for word in lemmatized_text:
    if word not in stopwords:
        processed_text.append(word)
print(processed_text)
```

```python
voca = list(set(processed_text))
print(voca)
```

```python
import numpy as np
import math
vocab_len = len(voca)

weighted_edge = np.zeros((vocab_len, vocab_len), dtype = np.float32)
score = np.zeros((vocab_len), dtype = np.float32)

window_size = 3
covered_coocurrences = []

for i in range(0, vocab_len):
    score[i] = 1
    for j in range(0,vocab_len):
        if j==i:
            continue
        else:
            for window_start in range(0,(len(processed_text)-window_size)):
                window_end = window_start+window_size
                window = processed_text[window_start: window_end]
                
                if (voca[i] in window) and (voca[j] in window):
                    index_of_i = window_start + window.index(voca[i])
                    index_of_j = window_start + window.index(voca[j])
                    print(i, voca[i], index_of_i, j, voca[j], index_of_j)
                    
                    if [index_of_i, index_of_j] not in covered_coocurrences:
                        weighted_edge[i][j] += 1/math.fabs(index_of_i-index_of_j)
                        print(i, j, weighted_edge[i][j])
                        covered_coocurrences.append([index_of_i,index_of_j])
```

```python
inout = np.zeros((vocab_len), dtype = np.float32)

for i in range(0, vocab_len):
    for j in range(0, vocab_len):
        inout[i]+=weighted_edge[i][j]
```

```python
import pandas as pd
inout2 = pd.DataFrame(weighted_edge, columns = range(len(weighted_edge))).sum()
inout2[0]
```

```python
inout[0]
```

```python
MAX_ITERATIONS = 50
d=0.85
threshold = 0.0001

for iter in range(0, MAX_ITERATIONS):
    prev_score = np.copy(score)
    
    for i in range(0, vocab_len):
        summation = 0
        for j in range(0, vocab_len):
            if weighted_edge[i][j] !=0:
                summation += (weighted_edge[i][j]/inout[j])*score[j]
        score[i] = (1-d) + d*summation
    
    if np.sum(np.fabs(prev_score-score)) <= threshold:
        break
for i in range(0,10):
    print("Score of "+voca[i]+":"+str(score[i]))
```

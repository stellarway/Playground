# 문장분리 실험

```python
from nltk.tokenize.punkt import PunktSentenceTokenizer, PunktParameters
```

```python
punkt_param = PunktParameters()
punkt_param.abbrev_types = set(['p.s','vs','etc','but'])
sent_splitter = PunktSentenceTokenizer(punkt_param)
```

```python
sample_sentence = df_total.iloc[1,1]
sample_sentence
```

```python
sent_splitter.tokenize(sample_sentence)
```

## map을 이용한 열 추가 생성

```python
def func(x):
    return sent_splitter.tokenize(x)
df_total['sentences'] = df_total['reviewText'].map(lambda x : func(x))
df_total
```

### ''는 None이 아니라 dropna()가 통하지 않으므로 플래트닝에서 오류를 발생시키는 []를 None값으로 치환

```python
def func(x):
    if x==[]:
        return None
    return len(x)
df_total['sentences_num'] = df_total['sentences'].map(lambda x : func(x))
df_total
```

### 아신을 문장갯수 만큼 반복하여 리스트 생성

```python
def func(x1,x2):
    return [str(x1)]*int(x2)
df_total['asin_rep'] = df_total.apply(lambda x : func(x['asin'], x['sentences_num']), axis=1)
df_total.head()
```

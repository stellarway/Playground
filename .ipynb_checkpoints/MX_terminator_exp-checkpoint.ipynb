```python
punkt_param = PunktParameters()
punkt_param.abbrev_types = set(['p.s','vs','etc', 'but'])
sent_splitter = PunktSentenceTokenizer(punkt_param)
```

```python
start = time.time() 
for idx, sent in enumerate(df['sentence']):
    feature =[eval('feat{}.search(sent).group()'.format(i)) for i in range(1,16)\
              if eval('feat{}.search(sent)'.format(i))!=None]
    if len(feature) >= 1:
        for i in feature:
            tempDf = df.iloc[idx]
            tempDf['feature'] = i
            featDf = featDf.append(tempDf)
            print("making feaDf Runtime: %0.2f sec"%(time.time() - start))
            print('-----{}번째 문장 \"{}\" feature 추가 완료-----'.format(idx,i), end = '\n\n')
print("making feaDf Runtime: %0.2f sec"%(time.time() - start))
print('-----모든 문장 feature 추가 완료-----')

```
